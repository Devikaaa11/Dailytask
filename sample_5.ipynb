```python
import pandas as pd
```

```python
sample=pd.read_excel('/content/Sample - Superstore.xls')
```

```python
sample.head()
```

```python
import matplotlib.pyplot as plt


sample['Order Date'] = pd.to_datetime(sample['Order Date'])
monthly_sales = sample.resample('M', on='Order Date')['Sales'].sum()

plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, label='Monthly Sales')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

# Bar plot

```python
city_sales = sample.groupby('City')['Sales'].sum()
top_5_cities = city_sales.sort_values(ascending=False).head(5)
plt.bar(top_5_cities.index,top_5_cities.values)
plt.xticks(rotation=90,fontsize=12)
plt.yticks(fontsize=12)
plt.title('Top 5 Cities by Sales', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7,c='k')

plt.show()
```

# Scatter plot


```python
plt.figure(figsize=(10, 6))
plt.scatter(sample['Sales'], sample['Profit'], color='red')
plt.title('Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

# Histogram

```python
plt.figure(figsize=(10, 6))
plt.hist(sample['Sales'],color='green', alpha=0.7)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

# Heatmap



```python
sample1=sample[['Profit','Sales','Postal Code']]
import seaborn as sns
sns.heatmap(sample1.corr())
```

# Pie chart

```python
region_sales = sample.groupby('Region')['Sales'].sum()

plt.figure(figsize=(10, 6))
region_sales.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
plt.title('Sales by Region')
plt.show()
```

Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
data1['income_binary']=data1['income'].apply(lambda x: 0 if x== '<50k' else 1)
data1
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
data1['age_group'] = data1['age'].apply(lambda x: 'Youth' if x < 30 else ('Adult' if x < 60 else 'Senior'))
data1
```

Rename the 'educational-num' column to 'education_num'.

```python
data1.rename(columns={'educational-num':'education_num'}, inplace=True)
data1
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
data1.isnull().sum()
```

```python
data1['workclass'].replace('?','unknown',inplace=True)
data1['occupation'].replace('?','unknown',inplace=True)
data1
```

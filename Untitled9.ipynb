```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Wholesale customers data.csv')
```

```python
from sklearn.cluster import KMeans
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i,n_init=10)
  km.fit(df)
  css=km.inertia_
  wcss.append(css)
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.title('elbow method')
plt.grid()
```

```python
km=KMeans(n_clusters=3,n_init=10)
km.fit(df)
```

```python
cluster_label=km.predict(df)
```

```python
cluster_label
```

```python
colors=['b','c','g']
for i in range(max(cluster_label+1)):
  cluster_data=df[cluster_label==i]
  plt.scatter(cluster_data['Fresh'],cluster_data['Delicassen'],c=colors[i],label=f'cluster{i}')
plt.xlabel('Fresh')
plt.ylabel('Delicassen')
plt.title('Fresh v/s Delicassen')
plt.legend()
plt.show()
```

```python
def predict_cluster(Channel,Region,fresh, milk,Grocery, Frozen,
       Detergents_Paper, Delicassen):

    new_data = [[Channel, Region,fresh, milk,Grocery, Frozen,
       Detergents_Paper, Delicassen]]
    cluster = km.predict(new_data)
    return cluster[0]
Channel=2
Region=3
fresh = 12669
milk = 9656
Grocery=4230
Frozen=5000
Detergents_Paper=6000
Delicassen=4000
predicted_cluster = predict_cluster(Channel,Region,fresh, milk,Grocery, Frozen,
       Detergents_Paper, Delicassen)
print(f"The predicted cluster for point (Fresh: {fresh}, Milk: {milk}, Grocery: {Grocery}, "
      f"Frozen: {Frozen}, Detergents_Paper: {Detergents_Paper}, Delicassen: {Delicassen}) "
      f"is: {predicted_cluster}")
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
features = df[['Channel','Region','Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
kmeans = KMeans(n_clusters=3, n_init=10)
df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
def predict_cluster1(Channel,Region,fresh, milk,Grocery, Frozen,
       Detergents_Paper, Delicassen):
    new_data1 = scaler.transform([[Channel, Region,fresh, milk,Grocery, Frozen,
       Detergents_Paper, Delicassen]])
    cluster = km.predict(new_data1)
    return cluster[0]


Channel=2
Region=3
fresh = 12669
milk = 9656
Grocery=4230
Frozen=5000
Detergents_Paper=6000
Delicassen=4000

predicted_cluster = predict_cluster1(Channel,Region,fresh, milk,Grocery, Frozen,
       Detergents_Paper, Delicassen)
print(f"The predicted cluster for point (Fresh: {fresh}, Milk: {milk}, Grocery: {Grocery}, "
      f"Frozen: {Frozen}, Detergents_Paper: {Detergents_Paper}, Delicassen: {Delicassen}) "
      f"is: {predicted_cluster}")
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
mergings=linkage(df,method='ward')
```

```python
plt.figure(figsize=(10,6))
dendrogram(mergings)
plt.show()
```

```python
cluster_label=fcluster(mergings,200,criterion='distance')
```

```python
for i in range(max(df['Cluster']) + 1):
    cluster_data = df[df['Cluster'] == i]
    plt.scatter(cluster_data['Fresh'], cluster_data['Milk'],
                c=colors[i],
                label=f'Cluster {i + 1}')

plt.xlabel('Fresh')
plt.ylabel('Milk')
plt.title('Fresh vs Milk')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil_avg=silhouette_score(df,cluster_label)
print(sil_avg)
```

```python
q1 = df['Vict Age'].quantile(0.25)
q3 = df['Vict Age'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['Vict Age'] < lower_bound) | (df['Vict Age'] > upper_bound)]['Vict Age']
print(outliers)
```

Q1 and Q3: These are the 25th and 75th percentiles, respectively.

IQR: This is the range between Q3 and Q1.
Bounds: Any value below the lower bound or above the upper bound is considered an outlier.


```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

correlation_matrix = df[numerical_columns].corr()

print(correlation_matrix)
```

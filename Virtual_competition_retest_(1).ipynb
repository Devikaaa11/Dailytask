**18.** Load the iris dataset.

````python
from sklearn.datasets import load_iris
iris=load_iris()
iris.data
````

**19.** Which species of iris has the highest average petal length?

````python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

avg_petal_length = df.groupby('species')['petal length (cm)'].mean()
h = avg_petal_length.idxmax()

print("Species with the highest average petal length:",h)
````

**20.** Draw a plot between sepal_length and sepal_width from the iris dataset.

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.show()
````
